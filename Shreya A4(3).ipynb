```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
from sklearn.manifold import MDS
import seaborn as sns
```

```python
# Load the icecream dataset
icecream_df = pd.read_csv( '/Users/shreyamishra/Desktop/icecream.csv')
```

```python
# Get the dimensions of the dataset
print(icecream_df.shape)
```

```python
# Get the column names of the dataset
print(icecream_df.columns)
```

```python
# Subset the dataset to exclude the 'Brand' column
ice = icecream_df.drop('Brand', axis=1)
```

```python
# Calculate the distance matrix
distance_matrix = distance.pdist(ice)
```

```python
# Convert the distance matrix to a square form
distance_matrix = distance.squareform(distance_matrix)
```

```python
# Perform MDS
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42)
mds_result = mds.fit_transform(distance_matrix)
```

```python
# Plot the MDS result
plt.figure(figsize=(8, 6))
plt.scatter(mds_result[:, 0], mds_result[:, 1])
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.title("MDS Plot")
plt.show()
```
